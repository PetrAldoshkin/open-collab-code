```python
import numpy as np

def str_to_array(s):
    # Разбиваем строку по пробелам и преобразуем в float
    return np.array([float(x) for x in s.split()])

# Применяем к столбцу embedding
df['embedding_array'] = df['embedding'].apply(str_to_array)
```

```python
import numpy as np

SIMILARITY_THRESHOLD = 0.09

def cosine_distance(a, b):
    # Проверка размера и типа
    a = np.array(a)
    b = np.array(b)
    if a.shape != b.shape:
        print(f"Warning: embedding shapes differ: {a.shape} vs {b.shape}")
        return 1.0
    denom = (np.linalg.norm(a) * np.linalg.norm(b))
    if denom == 0:
        return 1.0
    return 1 - np.dot(a, b) / denom

def find_similar_person(embedding, known_embeddings, threshold=SIMILARITY_THRESHOLD):
    for idx, known_emb in enumerate(known_embeddings):
        dist = cosine_distance(embedding, known_emb)
        if dist < threshold:
            return idx
    return None

# Предположим, у тебя df с колонкой 'embedding', где numpy array
persons_embeddings = []
person_ids = []
next_person_id = 0

for idx, row in df.iterrows():
    emb = row['embedding']
    if not isinstance(emb, np.ndarray):
        emb = np.array(emb)
    pid = find_similar_person(emb, persons_embeddings)
    if pid is None:
        # Новое лицо
        pid = next_person_id
        persons_embeddings.append(emb)
        next_person_id += 1
    person_ids.append(pid)

df['person_id'] = person_ids

print(f"Всего уникальных person_id: {next_person_id}")
print(df[['face_id', 'person_id']].head(10))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cosine
from sklearn.metrics import roc_curve, auc
import random

# Функция для вычисления расстояния косинуса
def cosine_distance(vec1, vec2):
    return cosine(vec1, vec2)

# Генерируем пары эмбеддингов для одного человека (positive pairs)
def generate_positive_pairs(df, max_pairs=5000):
    positive_pairs = []
    grouped = df.groupby('person_id')
    for person_id, group in grouped:
        embeddings = list(group['embedding'])
        if len(embeddings) < 2:
            continue
        # Перебираем пары внутри группы
        for i in range(len(embeddings)):
            for j in range(i+1, len(embeddings)):
                positive_pairs.append((embeddings[i], embeddings[j]))
                if len(positive_pairs) >= max_pairs:
                    return positive_pairs
    return positive_pairs

# Генерируем пары эмбеддингов для разных людей (negative pairs)
def generate_negative_pairs(df, max_pairs=5000):
    negative_pairs = []
    person_ids = df['person_id'].unique()
    for _ in range(max_pairs):
        p1, p2 = random.sample(list(person_ids), 2)
        e1 = df[df['person_id'] == p1]['embedding'].sample(1).values[0]
        e2 = df[df['person_id'] == p2]['embedding'].sample(1).values[0]
        negative_pairs.append((e1, e2))
    return negative_pairs

# Основной анализ
def analyze_threshold(df):
    print("Генерируем positive pairs...")
    positive_pairs = generate_positive_pairs(df)
    print(f"Позитивных пар: {len(positive_pairs)}")

    print("Генерируем negative pairs...")
    negative_pairs = generate_negative_pairs(df)
    print(f"Негативных пар: {len(negative_pairs)}")

    # Вычисляем расстояния
    pos_distances = [cosine_distance(e1, e2) for e1, e2 in positive_pairs]
    neg_distances = [cosine_distance(e1, e2) for e1, e2 in negative_pairs]

    # Визуализация
    plt.hist(pos_distances, bins=50, alpha=0.5, label='Same Person')
    plt.hist(neg_distances, bins=50, alpha=0.5, label='Different Persons')
    plt.xlabel('Cosine Distance')
    plt.ylabel('Count')
    plt.legend()
    plt.title('Distance Distribution')
    plt.show()

    # Создаём метки и объединяем данные
    y_true = np.array([1]*len(pos_distances) + [0]*len(neg_distances))
    y_scores = np.array(pos_distances + neg_distances)

    # Для ROC-кривой нам нужны похожие пары как низкие расстояния => обратим их
    y_scores = 1 - y_scores  # Чем больше - тем ближе

    fpr, tpr, thresholds = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.show()

    # Подбираем порог с максимальным TPR - FPR (Youden's J statistic)
    j_scores = tpr - fpr
    j_ordered = sorted(zip(j_scores, thresholds), reverse=True)
    best_j, best_thresh = j_ordered[0]

# Запускаем
analyze_threshold(df)
```

```python
import networkx as nx
from itertools import combinations

G = nx.Graph()

# Добавляем узлы (уникальные person_id)
G.add_nodes_from(df["person_id"].unique())

# Для каждой фотографии: находим, кто на ней присутствует
for pid in df["photo_id"].unique():
    persons = set(df[df["photo_id"] == pid]["person_id"])
    for u, v in combinations(persons, 2):
        if G.has_edge(u, v):
            G[u][v]["weight"] += 1
        else:
            G.add_edge(u, v, weight=1)

# Сохраняем ребра графа в таблицу
edges = [(u, v, G[u][v]["weight"]) for u, v in G.edges()]
edges_df = pd.DataFrame(edges, columns=["person_id_1", "person_id_2", "cooccurrence_count"])
edges_df.to_csv("graph_edges.csv", index=False)

print(f"Сохранено {len(edges_df)} ребер в файл graph_edges.csv.")
```

```python
edges_df = pd.read_csv("graph_edges.csv")

# 2. Строим граф
G = nx.Graph()
for _, row in edges_df.iterrows():
    G.add_edge(row["person_id_1"], row["person_id_2"], weight=row["cooccurrence_count"])

# 3. Настраиваем размеры узлов по степени (количеству связей)
node_size = [100 + 50 * G.degree(n) for n in G.nodes()]

# 4. Настраиваем толщину ребер по количеству встреч
edge_width = [G[u][v]['weight'] for u, v in G.edges()]

# 5. Расположение узлов (spring layout — красиво разносит)
pos = nx.spring_layout(G, k=0.5, iterations=50)

# 6. Рисуем
plt.figure(figsize=(12, 12))
nx.draw_networkx_nodes(G, pos, node_size=node_size, node_color='skyblue', alpha=0.8)
nx.draw_networkx_edges(G, pos, width=edge_width, edge_color='gray', alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=8)

plt.title("Социальный граф: встречи по фотографиям")
plt.axis("off")
plt.tight_layout()
plt.show()
```
